3. Si lo encuentra, debe devolver los datos del producto.

```python
def buscar_producto(inventario):
    try:
        codigo = int(input('Ingrese codigo del producto: '))
    except ValueError:
        print('El codigo no es valido')
    
    if codigo not in inventario.keys(): 
        raise KeyError('El producto no se encuentra en inventario')
    
    producto = inventario[codigo]
    print('Producto encontrado: ')
    print(producto)
    
    return producto
```

## **4. Estadísticas del inventario (4 pts)**

Crear una función `estadisticas(inventario)` que calcule:

1. Cantidad total de productos distintos.
2. Número total de unidades en stock.
3. Precio promedio de los productos.
4. Una lista de comprensión con los productos cuyo stock sea mayor a 5.
5. Un conjunto (set) con todas las categorías sin repetir.

> Debe manejar excepciones internas si el inventario está vacío.

```python
def estadisticas(inventario):
    print(f'Cantidad de productos distintos: {len(inventario)}')
    stockTotal = 0
    precios = []
    for producto in inventario.values():
        stockTotal += producto['stock']
        precios.append(producto['precio'])
    print(f'Numero total de unidades en stock: {stockTotal}')
    if precios:
        precioPromedio = sum(precios) / len(precios)
    else:
        precioPromedio = 0
    print(f"Precio promedio: {precioPromedio}")
    #Stock mayor a 5
    productosEnStock = [v['nombre'] for v in inventario.values() if v['stock'] > 5]
    print(f'Productos con stock mayor a 5: {productosEnStock}')
    #Categorias no repetidas
    categorias = set([v['categoria'] for v in inventario.values()])
    print(f'Categorias: {categorias}')
```
